## Solution 2 — Data Cleaning

```python
customers = pd.DataFrame({
    'name':        ['  Alice ', 'Bob', '  Carol', 'Dave', 'Bob'],
    'email':       ['alice@x.com ', 'bob@x.com', 'carol@x.com', ' dave@x.com', 'bob@x.com'],
    'age':         [28, None, 34, 41, None],
    'signup_date': ['2023-01-15', '2023-03-22', '2023-06-01', '2022-11-30', '2023-03-22'],
})
customers['name'] = customers['name'].str.strip()
customers['email'] = customers['email'].str.strip()
customers['signup_date'] = pd.to_datetime(customers['signup_date'])
customers['age'] = customers['age'].fillna(customers['age'].median())
customers = customers.drop_duplicates().reset_index(drop=True)
print(customers)
```

## Solution 3 — GroupBy & Aggregation

```python
employees = pd.DataFrame({
    'name':   ['Alice','Bob','Carol','Dave','Eve','Frank','Grace','Hank','Ivy'],
    'dept':   ['Eng','Eng','Sales','Sales','HR','HR','Eng','Sales','HR'],
    'level':  ['Senior','Junior','Senior','Junior','Senior','Junior','Mid','Mid','Junior'],
    'salary': [120000,75000,95000,65000,70000,55000,90000,72000,60000],
})
print('1. Avg salary per dept:')
print(employees.groupby('dept')['salary'].mean())

print('2. Full stats:')
print(employees.groupby('dept')['salary'].agg(['count','min','max','sum']))

best_dept = employees.groupby('dept')['salary'].mean().idxmax()
print('3. Highest avg salary dept:', best_dept)

employees['salary_rank'] = employees.groupby('dept')['salary'].rank(ascending=False, method='min').astype(int)
print(employees[['name','dept','salary','salary_rank']])
```

## Solution 4 — Merge & Join

```python
orders = pd.DataFrame({'order_id':[1,2,3,4,5],'cust_id':[101,102,101,103,102],'prod_id':['P1','P2','P3','P1','P2'],'qty':[2,1,3,1,2]})
customers = pd.DataFrame({'cust_id':[101,102,103,104],'name':['Alice','Bob','Carol','Dave'],'city':['NY','LA','NY','Chicago']})
products = pd.DataFrame({'prod_id':['P1','P2','P3'],'name':['Widget','Gadget','Doohickey'],'price':[9.99,24.99,4.99]})

merged = orders.merge(customers, on='cust_id', how='inner')
merged = merged.merge(products, on='prod_id', how='left', suffixes=('','_product'))
print('Merged:')
print(merged)

no_orders = customers[~customers['cust_id'].isin(orders['cust_id'])]
print('No orders:', no_orders[['cust_id','name']].to_string(index=False))

merged['spend'] = merged['qty'] * merged['price']
print('Total spend per customer:')
print(merged.groupby('name')['spend'].sum().sort_values(ascending=False))
```

## Solution 8 — Apply & Custom Functions

```python
np.random.seed(7)
n = 300
loans = pd.DataFrame({'applicant_id':range(1,n+1),'income':np.random.randint(25000,150000,n),'debt':np.random.randint(5000,80000,n),'credit_score':np.random.randint(300,850,n)})
loans['debt_ratio'] = (loans['debt'] / loans['income']).round(2)
loans['risk_score'] = ((loans['credit_score'] - 300) / (850 - 300) * 100).round(1)
loans['risk_category'] = pd.cut(loans['risk_score'], bins=[-1,40,70,101], labels=['High','Medium','Low'])
loans['approved'] = (loans['debt_ratio'] < 0.4) & (loans['risk_score'] >= 60)
approval_rate = loans.groupby('risk_category', observed=True)['approved'].mean().round(3)
print('Approval rate per risk category:')
print(approval_rate)
```

## Solution 9 — Window Functions

```python
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=120, freq='B')
returns = np.random.normal(0.0005, 0.015, 120)
prices = 100 * np.cumprod(1 + returns)
stock = pd.DataFrame({'close': prices.round(2)}, index=dates)
stock['5d_ma'] = stock['close'].rolling(5).mean().round(2)
stock['20d_ma'] = stock['close'].rolling(20).mean().round(2)
stock['pct_change'] = stock['close'].pct_change().round(4)
stock['big_move'] = stock['pct_change'].abs() > 0.03
stock['cum_return'] = (stock['close'] / stock['close'].iloc[0]).round(4)
rolling_max = stock['close'].cummax()
stock['drawdown'] = ((stock['close'] - rolling_max) / rolling_max).round(4)
print('Max drawdown:', stock['drawdown'].min())
prev_above = stock['5d_ma'] > stock['20d_ma']
curr_above = stock['5d_ma'] > stock['20d_ma']
stock['signal'] = 'HOLD'
cross_above = (~prev_above.shift(1).fillna(False)) & curr_above
cross_below = prev_above.shift(1).fillna(False) & (~curr_above)
stock.loc[cross_above, 'signal'] = 'BUY'
stock.loc[cross_below, 'signal'] = 'SELL'
print(stock[['close','5d_ma','20d_ma','signal']].dropna().tail(10))
```

## Solution 10 — E-Commerce Mini Project

```python
from io import StringIO
np.random.seed(42)
n = 500
ecommerce = pd.DataFrame({
    'order_id':    range(1001, 1001+n),
    'customer_id': np.random.choice([f'C{i:03d}' for i in range(1,81)] + [None], n, p=[1/81]*80 + [1/81]),
    'order_date':  pd.date_range('2023-01-01', periods=n, freq='14H').strftime('%Y-%m-%d'),
    'category':    np.random.choice(['Electronics','Clothing','Books','Home','Sports'], n),
    'quantity':    np.random.randint(1, 6, n),
    'unit_price':  np.random.choice([9.99,19.99,49.99,99.99,199.99], n),
    'discount':    np.random.choice([0, 0, 0, 0.1, 0.2, None], n),
})
print('Shape:', ecommerce.shape)
ecommerce['order_date'] = pd.to_datetime(ecommerce['order_date'])
ecommerce['discount'] = ecommerce['discount'].fillna(0)
ecommerce = ecommerce.dropna(subset=['customer_id'])
ecommerce['net_revenue'] = (ecommerce['quantity'] * ecommerce['unit_price'] * (1 - ecommerce['discount'])).round(2)
monthly = ecommerce.groupby(ecommerce['order_date'].dt.to_period('M'))['net_revenue'].sum()
print('Monthly revenue:')
print(monthly)
top5 = ecommerce.groupby('customer_id')['net_revenue'].sum().nlargest(5)
print('Top 5 customers:')
print(top5)
cat = ecommerce.groupby('category').agg(total_orders=('order_id','count'), avg_order_value=('net_revenue','mean')).round(2)
print('Category performance:')
print(cat)
first_order = ecommerce.groupby('customer_id')['order_date'].min().dt.to_period('M').rename('first_order_month')
ecommerce = ecommerce.join(first_order, on='customer_id')
order_counts = ecommerce.groupby('customer_id')['order_id'].count()
ecommerce['is_repeat'] = ecommerce['customer_id'].map(order_counts) > 1
print('Repeat customers:', ecommerce['is_repeat'].sum())
summary = cat.reset_index()
buf = StringIO()
summary.to_csv(buf, index=False)
print('CSV preview:')
print(buf.getvalue())
```
