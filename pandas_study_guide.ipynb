## 2. Series <a id='2'></a>
A `Series` is a 1-D labeled array. Think of it as a single column of a spreadsheet.

```python
# --- Creating a Series ---

# From a list
s1 = pd.Series([10, 20, 30, 40, 50])
print("From list:")
print(s1)

# From a list with custom index
s2 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print("\nCustom index:")
print(s2)

# From a dictionary
s3 = pd.Series({'apples': 5, 'bananas': 3, 'cherries': 8})
print("\nFrom dict:")
print(s3)
```

```python
# --- Accessing Series elements ---

print("Element at index 'b':", s2['b'])
print("Element at position 1:", s2.iloc[1])
print("First two elements:\n", s2.iloc[:2])

# Series attributes
print("\nValues:", s2.values)      # numpy array
print("Index:", s2.index.tolist())
print("dtype:", s2.dtype)
print("Name:", s2.name)
```

```python
# --- Series arithmetic ---
s = pd.Series([1, 2, 3, 4, 5])

print("Original:", s.values)
print("Multiply by 2:", (s * 2).values)
print("Add 10:", (s + 10).values)
print("Squared:", (s ** 2).values)
print("Sum:", s.sum())
print("Mean:", s.mean())
print("Max:", s.max())
```

## 7. Filtering Data <a id='7'></a>

## 9. Handling Missing Data <a id='9'></a>

```python
# Create a DataFrame with missing values
df_miss = pd.DataFrame({
    'A': [1, 2, np.nan, 4, np.nan],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': ['x', 'y', None, 'w', 'v'],
    'D': [10, 20, 30, 40, 50]
})
df_miss
```

```python
# --- Detect missing values ---
print("isnull():")
print(df_miss.isnull())

print("\nCount of nulls per column:")
print(df_miss.isnull().sum())

print("\nAny null per column:")
print(df_miss.isnull().any())
```

```python
# --- Drop missing values ---
print("Drop rows with ANY null:")
print(df_miss.dropna())

print("\nDrop rows where ALL values are null:")
print(df_miss.dropna(how='all'))

print("\nDrop rows with null in column 'A':")
print(df_miss.dropna(subset=['A']))
```

```python
# --- Fill missing values ---
print("Fill all nulls with 0:")
print(df_miss.fillna(0))

print("\nFill column A with its mean:")
df_filled = df_miss.copy()
df_filled['A'] = df_filled['A'].fillna(df_filled['A'].mean())
print(df_filled)

print("\nForward fill (propagate last valid value):")
print(df_miss.ffill())

print("\nBackward fill:")
print(df_miss.bfill())
```

## 12. GroupBy & Aggregation <a id='12'></a>

```python
sales = pd.DataFrame({
    'Region':   ['East', 'West', 'East', 'West', 'North', 'North', 'East', 'West'],
    'Product':  ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
    'Sales':    [200, 300, 150, 250, 180, 220, 210, 190],
    'Quantity': [10, 15, 8, 12, 9, 11, 11, 10],
    'Month':    ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'Mar']
})
sales
```

```python
# --- Basic groupby ---
print("Total Sales by Region:")
print(sales.groupby('Region')['Sales'].sum())

print("\nMean Sales by Region:")
print(sales.groupby('Region')['Sales'].mean().round(1))
```

```python
# --- Multiple aggregations with agg() ---
print("Multiple stats per Region:")
print(sales.groupby('Region')['Sales'].agg(['sum', 'mean', 'min', 'max', 'count']))

print("\nCustom agg per column:")
print(sales.groupby('Region').agg({'Sales': ['sum', 'mean'], 'Quantity': 'sum'}))
```

```python
# --- Group by multiple columns ---
print("Sales by Region and Product:")
print(sales.groupby(['Region', 'Product'])['Sales'].sum().unstack())
```

```python
# --- transform: return same-shape result ---
# Useful for adding group-level stats back to the original df
sales['RegionTotalSales'] = sales.groupby('Region')['Sales'].transform('sum')
sales['SalesShare%'] = (sales['Sales'] / sales['RegionTotalSales'] * 100).round(1)
print(sales[['Region', 'Sales', 'RegionTotalSales', 'SalesShare%']])
```

```python
# --- filter: keep groups satisfying a condition ---
# Keep only regions where total sales > 500
big_regions = sales.groupby('Region').filter(lambda g: g['Sales'].sum() > 500)
print("Regions with total sales > 500:")
print(big_regions[['Region', 'Sales']])
```

## 13. Merging, Joining & Concatenating <a id='13'></a>

```python
left = pd.DataFrame({
    'id':   [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'dept': ['Eng', 'Mkt', 'Eng', 'HR']
})

right = pd.DataFrame({
    'id':     [2, 3, 4, 5],
    'salary': [80000, 90000, 72000, 65000],
    'city':   ['LA', 'Chicago', 'NYC', 'Boston']
})

print("Left:\n", left)
print("\nRight:\n", right)
```

```python
# --- merge (like SQL joins) ---
print("INNER JOIN (only matching ids):")
print(pd.merge(left, right, on='id', how='inner'))

print("\nLEFT JOIN (all from left):")
print(pd.merge(left, right, on='id', how='left'))

print("\nOUTER JOIN (all records):")
print(pd.merge(left, right, on='id', how='outer'))
```

```python
# --- merge on different column names ---
emp  = pd.DataFrame({'emp_id': [1,2,3], 'name': ['X','Y','Z']})
dept = pd.DataFrame({'department_emp_id': [1,2,3], 'dept': ['Eng','HR','Mkt']})
print(pd.merge(emp, dept, left_on='emp_id', right_on='department_emp_id'))
```

```python
# --- concat: stack DataFrames ---
df_a = pd.DataFrame({'x': [1,2], 'y': [3,4]})
df_b = pd.DataFrame({'x': [5,6], 'y': [7,8]})
df_c = pd.DataFrame({'z': [9,10]})

print("Vertical concat (axis=0):")
print(pd.concat([df_a, df_b], ignore_index=True))

print("\nHorizontal concat (axis=1):")
print(pd.concat([df_a, df_c], axis=1))
```

## 14. Pivot Tables & Crosstab <a id='14'></a>

```python
# --- pivot_table ---
print("Pivot: Average Sales by Region & Product:")
pivot = pd.pivot_table(sales, values='Sales', index='Region',
                       columns='Product', aggfunc='mean', fill_value=0)
print(pivot)

print("\nPivot with margins (totals):")
print(pd.pivot_table(sales, values='Sales', index='Region',
                     columns='Product', aggfunc='sum',
                     fill_value=0, margins=True))
```

## 15. String Operations <a id='15'></a>

```python
s_str = pd.Series(['  Hello World  ', 'python PANDAS', 'Data Science 101', 'foo@bar.com'])

print("lower:     ", s_str.str.lower().tolist())
print("upper:     ", s_str.str.upper().tolist())
print("strip:     ", s_str.str.strip().tolist())
print("len:       ", s_str.str.len().tolist())
print("contains 'data':", s_str.str.contains('data', case=False).tolist())
print("startswith 'py':", s_str.str.startswith('py').tolist())
print("replace:   ", s_str.str.replace('o', '0', regex=False).tolist())
```

## 16. DateTime Operations <a id='16'></a>

```python
dates = pd.to_datetime(['2024-01-15', '2024-03-22', '2024-07-04',
                        '2024-11-11', '2025-02-28'])
ds = pd.Series(dates)

print("Dates:    ", ds.tolist())
print("Year:     ", ds.dt.year.tolist())
print("Month:    ", ds.dt.month.tolist())
print("Day:      ", ds.dt.day.tolist())
print("Weekday:  ", ds.dt.day_name().tolist())
print("Week num: ", ds.dt.isocalendar().week.tolist())
print("Quarter:  ", ds.dt.quarter.tolist())
```

```python
# --- Date arithmetic ---
today = pd.Timestamp('2025-01-01')
print("Today:", today)
print("+ 30 days:", today + pd.Timedelta(days=30))
print("Days between dates:", (ds.max() - ds.min()).days)

# date_range
dr = pd.date_range(start='2024-01-01', end='2024-01-07', freq='D')
print("\nDate range:", dr.tolist())

# Business days
bdays = pd.date_range('2024-01-01', periods=5, freq='B')
print("Business days:", bdays.tolist())
```

```python
# --- Resample time series ---
ts = pd.Series(
    np.random.randint(100, 500, 90),
    index=pd.date_range('2024-01-01', periods=90, freq='D')
)

print("Daily → Monthly sum:")
print(ts.resample('ME').sum())

print("\nDaily → Weekly mean:")
print(ts.resample('W').mean().head())
```

## 17. Apply, Map & Lambda <a id='17'></a>

```python
# --- apply on a Series ---
def classify_age(age):
    if age < 25:   return 'Young'
    elif age < 35: return 'Middle'
    else:          return 'Senior'

df['AgeGroup'] = df['Age'].apply(classify_age)
print(df[['Name', 'Age', 'AgeGroup']])
```

```python
# --- apply with lambda ---
df['SalaryK'] = df['Salary'].apply(lambda x: f'${x/1000:.0f}K')
print(df[['Name', 'Salary', 'SalaryK']])
```

```python
# --- apply on a DataFrame (row-wise) ---
df['Profile'] = df.apply(
    lambda row: f"{row['Name']} ({row['Age']}) - {row['City']}",
    axis=1
)
print(df[['Profile']])
```

```python
# --- map: element-wise transformation using dict ---
city_region = {'NYC': 'Northeast', 'LA': 'West', 'Chicago': 'Midwest',
               'Houston': 'South', 'Boston': 'Northeast'}
df['Region'] = df['City'].map(city_region)
print(df[['Name', 'City', 'Region']])
```

```python
# --- np.where: vectorized if-else (faster than apply) ---
df['HighEarner'] = np.where(df['Salary'] >= 80000, 'Yes', 'No')
print(df[['Name', 'Salary', 'HighEarner']])
```

## 18. Reshaping Data <a id='18'></a>

```python
# --- melt: wide → long format ---
wide = pd.DataFrame({
    'Name':  ['Alice', 'Bob', 'Charlie'],
    'Math':  [90, 75, 85],
    'English': [80, 88, 72],
    'Science': [95, 70, 88]
})
print("Wide format:")
print(wide)

long = pd.melt(wide, id_vars=['Name'], var_name='Subject', value_name='Score')
print("\nLong format (melted):")
print(long)
```

```python
# --- pivot: long → wide ---
back_to_wide = long.pivot(index='Name', columns='Subject', values='Score')
back_to_wide.columns.name = None
print("Back to wide:")
print(back_to_wide.reset_index())
```

```python
# --- stack / unstack ---
df_multi = pd.DataFrame(
    np.arange(12).reshape(3, 4),
    index=['r1', 'r2', 'r3'],
    columns=pd.MultiIndex.from_tuples([('A','x'),('A','y'),('B','x'),('B','y')])
)
print("Multi-column DataFrame:")
print(df_multi)

print("\nStack (columns → rows):")
print(df_multi.stack(level=0))
```

## 19. Statistical Functions <a id='19'></a>

```python
numeric_df = df[['Age', 'Salary', 'Score']].copy()

print("=== Basic Stats ===")
print("Mean:\n",   numeric_df.mean())
print("\nMedian:\n", numeric_df.median())
print("\nStd:\n",    numeric_df.std())
print("\nVar:\n",    numeric_df.var())
print("\nMin:\n",    numeric_df.min())
print("\nMax:\n",    numeric_df.max())
```

```python
print("=== Correlation Matrix ===")
print(numeric_df.corr().round(3))

print("\n=== Covariance Matrix ===")
print(numeric_df.cov().round(1))
```

```python
# --- Cumulative stats ---
s = pd.Series([3, 1, 4, 1, 5, 9, 2, 6])
print("Values:   ", s.tolist())
print("cumsum:   ", s.cumsum().tolist())
print("cumprod:  ", s.cumprod().tolist())
print("cummax:   ", s.cummax().tolist())
print("cummin:   ", s.cummin().tolist())
print("pct_change:", s.pct_change().round(2).tolist())
```

```python
# --- Rolling windows ---
prices = pd.Series([100, 102, 98, 105, 103, 108, 110, 107, 112, 115])
print("3-day rolling mean:", prices.rolling(3).mean().round(2).tolist())
print("3-day rolling std: ", prices.rolling(3).std().round(2).tolist())
```

```python
# --- Quantiles & percentile ranks ---
print("Quantiles:")
print(numeric_df['Salary'].quantile([0.25, 0.5, 0.75]))

print("\nPercentile rank:")
print(numeric_df['Salary'].rank(pct=True).round(2))
```

## Quick Reference Cheatsheet

| Task | Code |
|------|------|
| Create DataFrame | `pd.DataFrame({'col': [...]})` |
| Read CSV | `pd.read_csv('file.csv')` |
| Select column | `df['col']` or `df.col` |
| Select rows | `df.loc[0:5]` / `df.iloc[0:5]` |
| Filter | `df[df['col'] > 10]` |
| Group & Aggregate | `df.groupby('col').agg(...)` |
| Sort | `df.sort_values('col', ascending=False)` |
| Missing values | `df.fillna(0)` / `df.dropna()` |
| Merge | `pd.merge(left, right, on='id', how='inner')` |
| Stack DataFrames | `pd.concat([df1, df2])` |
| Wide → Long | `pd.melt(df, id_vars=[...])` |

## 23. assign(), pipe() & Method Chaining <a id='23'></a>

```python
df_base = pd.DataFrame({
    'Name':   ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age':    [25, 30, 35, 28, 22],
    'Salary': [70000, 85000, 90000, 72000, 65000],
    'Score':  [88.5, 92.0, 78.3, 95.1, 85.6]
})

# --- assign: add/overwrite columns without mutating the original ---
result = df_base.assign(
    Tax      = lambda x: x['Salary'] * 0.2,
    NetPay   = lambda x: x['Salary'] - x['Salary'] * 0.2,
    Grade    = lambda x: pd.cut(x['Score'], bins=[0,75,85,100],
                                 labels=['C','B','A'])
)
print(result)
```

```python
# --- pipe: pass DataFrame into a custom function ---
def normalize_salary(df, col='Salary'):
    df = df.copy()
    df[col + '_norm'] = (df[col] - df[col].mean()) / df[col].std()
    return df

def flag_high_score(df, threshold=88):
    df = df.copy()
    df['TopPerformer'] = df['Score'] > threshold
    return df

# Method chaining using pipe
result = (
    df_base
    .query('Age >= 25')
    .assign(SalaryK=lambda x: (x['Salary'] / 1000).round(1))
    .pipe(normalize_salary)
    .pipe(flag_high_score, threshold=88)
    .sort_values('Salary', ascending=False)
    .reset_index(drop=True)
)
print(result)
```

## 27. Window Functions (expanding, ewm) <a id='27'></a>

```python
prices = pd.Series([10, 12, 11, 14, 13, 16, 15, 18, 17, 20],
                   name='Price')

# --- expanding: growing window from start ---
print("Expanding mean (cumulative average):")
print(prices.expanding().mean().round(2))

print("\nExpanding min / max:")
exp = pd.DataFrame({
    'Price':    prices,
    'exp_mean': prices.expanding().mean().round(2),
    'exp_min':  prices.expanding().min(),
    'exp_max':  prices.expanding().max(),
})
print(exp)
```

```python
# --- ewm: Exponentially Weighted Moving Average ---
# span=N: weight decays like a rolling window of size N
# alpha: smoothing factor (0 < alpha <= 1); higher = more weight on recent

print("EWM with span=3:")
print(prices.ewm(span=3).mean().round(3))

print("\nEWM with alpha=0.3 (explicit smoothing factor):")
print(prices.ewm(alpha=0.3).mean().round(3))

# Compare rolling vs EWM
compare = pd.DataFrame({
    'Price':      prices,
    'Roll3':      prices.rolling(3).mean().round(2),
    'EWM_span3':  prices.ewm(span=3).mean().round(2),
})
print("\nRolling 3 vs EWM span=3:")
print(compare)
```

## 29. Memory Optimization <a id='29'></a>

```python
# Create a large-ish DataFrame to demonstrate memory
np.random.seed(0)
n = 100_000

df_mem = pd.DataFrame({
    'id':       np.random.randint(0, 10000, n),
    'category': np.random.choice(['A','B','C','D','E'], n),
    'value':    np.random.uniform(0, 1000, n),
    'flag':     np.random.randint(0, 2, n),
    'big_int':  np.random.randint(0, 100, n).astype('int64'),
})

def mem_report(df):
    mb = df.memory_usage(deep=True).sum() / 1e6
    print(f"Total memory: {mb:.2f} MB")
    print(df.dtypes)

print("=== Before optimization ===")
mem_report(df_mem)
```

```python
df_opt = df_mem.copy()

# Downcast numeric types
df_opt['id']       = pd.to_numeric(df_opt['id'], downcast='unsigned')   # int64→uint16
df_opt['flag']     = df_opt['flag'].astype('int8')
df_opt['big_int']  = df_opt['big_int'].astype('int8')
df_opt['value']    = df_opt['value'].astype('float32')

# Categorical for low-cardinality strings
df_opt['category'] = df_opt['category'].astype('category')

print("=== After optimization ===")
mem_report(df_opt)

before = df_mem.memory_usage(deep=True).sum()
after  = df_opt.memory_usage(deep=True).sum()
print(f"\nSavings: {(1 - after/before)*100:.1f}%")
```

```python
# --- Weighted sampling (higher salary = more likely to be picked) ---
print("Weighted sample by Salary:")
print(df_s.sample(n=4, weights='Salary', random_state=42))

# --- Stratified sample: same proportion per group ---
def stratified_sample(df, group_col, frac=0.5, random_state=42):
    return df.groupby(group_col, group_keys=False).apply(
        lambda x: x.sample(frac=frac, random_state=random_state)
    )

print("\nStratified sample (50% per Dept):")
print(stratified_sample(df_s, 'Dept'))

# --- Shuffle the DataFrame ---
print("\nShuffled:")
print(df_s.sample(frac=1, random_state=42).reset_index(drop=True))
```

## 31. shift() & diff() for Time Series <a id='31'></a>

```python
stock = pd.DataFrame({
    'Date':  pd.date_range('2024-01-01', periods=8, freq='B'),
    'Close': [150.0, 152.5, 149.0, 155.0, 157.3, 153.8, 160.0, 162.1]
}).set_index('Date')

# --- shift: lag / lead ---
stock['Prev_Close']  = stock['Close'].shift(1)   # lag 1 day
stock['Next_Close']  = stock['Close'].shift(-1)  # lead 1 day
print("Shift:")
print(stock)
```

```python
# --- diff: simple difference ---
stock['Change']    = stock['Close'].diff(1)        # day-over-day change
stock['2day_diff'] = stock['Close'].diff(2)        # 2-period difference

# Percent change
stock['Pct_Change'] = stock['Close'].pct_change().mul(100).round(2)

print("Differences & % change:")
print(stock[['Close','Change','Pct_Change']])

# Detect up/down days
stock['Direction'] = np.where(stock['Change'] > 0, 'UP',
                     np.where(stock['Change'] < 0, 'DOWN', 'FLAT'))
print("\nDirections:")
print(stock[['Close','Change','Direction']])
```

### 17 · Tiered Discount Engine (Apply / Map)

```python
import numpy as np
orders = pd.DataFrame({
    "order_id":[1001,1002,1003], "customer":["Alice","Bob","Charlie"],
    "subtotal":[650., 320., 95.], "vip":[True, False, False],
})
def rate(row):
    base = 0.15 if row["subtotal"]>=500 else (0.10 if row["subtotal"]>=200 else 0.05)
    return base + (0.05 if row["vip"] else 0)
orders["discount_pct"]  = orders.apply(rate, axis=1)
orders["final_price"]   = (orders["subtotal"]*(1-orders["discount_pct"])).round(2)
print(orders[["customer","subtotal","discount_pct","final_price"]].to_string(index=False))
```

### 23 · ETL Pipeline (assign / pipe)

```python
import numpy as np
raw = pd.DataFrame({
    "customer":["  alice  ","BOB","charlie"], "amount":[250.,-99.,430.],
    "date":["2024-01-15","2024-01-16","2024-01-16"], "status":["completed","cancelled","completed"],
})
def clean(df):
    df=df.copy(); df["customer"]=df["customer"].str.strip().str.title(); return df
def valid(df): return df[(df["amount"]>0)]
result = (raw.pipe(clean).pipe(valid)
            .assign(date=lambda x:pd.to_datetime(x["date"]),
                    revenue=lambda x:x["amount"])
            .query("status=='completed'"))
print(result.to_string(index=False))
```

### 24 · Fast P&L Ratios (eval)

```python
import numpy as np
np.random.seed(0)
fin = pd.DataFrame({"revenue":np.random.uniform(1e6,50e6,5),"cogs_pct":np.random.uniform(.3,.6,5),"opex_pct":np.random.uniform(.1,.2,5)})
fin.eval("cogs        = revenue * cogs_pct",inplace=True)
fin.eval("gross       = revenue - cogs",    inplace=True)
fin.eval("opex        = revenue * opex_pct",inplace=True)
fin.eval("ebit        = gross - opex",      inplace=True)
fin.eval("margin      = gross / revenue",   inplace=True)
print(fin[["revenue","gross","ebit","margin"]].round(2))
```

### 27 · Stock Technical Indicators (ewm / expanding)

```python
import numpy as np
np.random.seed(42)
p = pd.Series(100+np.cumsum(np.random.normal(.2,2,30)),
              index=pd.date_range("2024-01-01",periods=30,freq="B"),name="Close")
ta = pd.DataFrame({"Close":p, "SMA_10":p.rolling(10).mean(),
                   "EMA_12":p.ewm(span=12).mean(), "ATH":p.expanding().max()})
ta["BB_upper"] = ta["SMA_10"] + 2*p.rolling(10).std()
ta["BB_lower"] = ta["SMA_10"] - 2*p.rolling(10).std()
ta["buy_sig"]  = p < ta["BB_lower"]
print(ta.tail(8).round(2))
print("Buy signals:", ta["buy_sig"].sum())
```

### 29 · Memory Optimisation

```python
import numpy as np
np.random.seed(0); n=200_000
txn = pd.DataFrame({"id":np.arange(n),"cat":np.random.choice(["A","B","C"],n),
                    "val":np.random.uniform(1,500,n),"flag":np.random.randint(0,2,n)})
before = txn.memory_usage(deep=True).sum()/1e6
txn["id"]  = txn["id"].astype("uint32")
txn["cat"] = txn["cat"].astype("category")
txn["val"] = txn["val"].astype("float32")
txn["flag"]= txn["flag"].astype("int8")
after = txn.memory_usage(deep=True).sum()/1e6
print(f"Before: {before:.1f}MB  →  After: {after:.1f}MB  (saved {(1-after/before)*100:.0f}%)")
```

### 30 · Stratified Train/Test Split (Sampling)

### 31 · Portfolio Returns & Drawdown (shift / diff)

```python
import numpy as np
np.random.seed(1)
nav = pd.Series([10000]+list(10000*np.cumprod(1+np.random.normal(.0005,.015,29))),
                index=pd.date_range("2024-01-02",periods=30,freq="B"),name="NAV")
pf = pd.DataFrame({"NAV":nav})
pf["daily_ret"]  = pf["NAV"].pct_change()
pf["cum_ret_%"]  = (pf["NAV"]/pf["NAV"].iloc[0]-1)*100
pf["rolling_max"]= pf["NAV"].expanding().max()
pf["drawdown_%"] = (pf["NAV"]-pf["rolling_max"])/pf["rolling_max"]*100
sharpe = pf["daily_ret"].mean()/pf["daily_ret"].std()*252**0.5
print(pf.tail(5).round(3))
print(f"Sharpe: {sharpe:.2f}  |  Max DD: {pf['drawdown_%'].min():.2f}%")
```
